## Operations: *, @, and np.matmul / np.dot

Elementwise and matrix multiplication are treated differently in Numpy.

```
import numpy as np


A = np.array([[1,2],[3,4]])
B = np.array([[10,20],[30,40]])
v = np.array([1,2])


print("Elementwise A * B:\n", A * B)
print("Matrix multiply A @ B:\n", A @ B)
print("np.matmul(A, B):\n", np.matmul(A, B))
print("np.dot(A, B):\n", np.dot(A, B))
print("np.dot(v, v) (inner product):", np.dot(v, v))


# Higher-dimensional example showing np.matmul broadcasting on leading dims
X = np.random.randn(2, 3, 4)
Y = np.random.randn(2, 4, 5)
print('X.shape, Y.shape =', X.shape, Y.shape)
print('np.matmul(X, Y).shape =', np.matmul(X, Y).shape) # (2,3,5)
```

## Vectorization and Performance

_Vectorization_ means applying operations to whole arrays at once, rather than element by element.

```
import math
import numpy as np


xs = np.linspace(0, 10, 1_000_000)


# Loop (slow) - run this in its own cell with %%timeit
# %%timeit
# ys_loop = np.zeros_like(xs)
# for i in range(len(xs)):
# ys_loop[i] = math.sin(xs[i])


# Vectorized (fast) - run this in its own cell with %%timeit
# %%timeit
# y_vec = np.sin(xs)
```

```
import numpy as np


def total_energy(k, m, x, v):
"""Return array of total energies for harmonic oscillator.


k: spring constant (scalar)
m: mass (scalar)
x: displacement array (or scalar)
v: velocity array (same shape as x or broadcastable)
"""
x = np.asarray(x)
v = np.asarray(v)
return 0.5 * m * v**2 + 0.5 * k * x**2


# Example usage
k = 1.0
m = 1.0
x = np.linspace(-1, 1, 9)
v = np.sin(x) # arbitrary example velocities
E = total_energy(k, m, x, v)
E
```

## Eigenvalues and Eigenvectors

For a matrix $M$, eigenvalues $\lambda$ and eigenvectors $v$ satisfy:
$$ M v = \lambda v $$

In physics, these often realte to characteristic and measurable physical quantities:
* Oscillators → $\lambda$ relates to squared frequencies
* Quantum mechanics → $\lambda$ could be the energy eigenstates
* Rotations → $\lambda$ could define a principal axes

### Here is an example matrix

```
K = np.array([[ 1,-1, 0],
[-1, 2,-1],
[ 0,-1, 1]])


vals, vecs = np.linalg.eigh(K)
print('Eigenvalues:', vals)
print('Eigenvectors (columns):\n', vecs)
```

```
for i, lam in enumerate(vals):
v = vecs[:, i]
residual = K @ v - lam * v
print(f'i={i}, λ={lam:.6g}, ||K v - λ v|| = {np.linalg.norm(residual):.2e}')


# Interpretation: vecs[:, i] is the eigenvector corresponding to vals[i].
```

```
K = np.array([[ 1,-1, 0]
             ,[-1, 2,-1]
             ,[ 0,-1, 1] ])
vals, vecs = np.linalg.eigh(K)
vals, vecs
```

## Quadratic Form: Diatomic Molecule Model

We can model a diatomic molecule (O$_2$ or N$_2$, for example) as two masses connected by a spring. Motion is restricted to the
bond axis.

The (quadratic) potential energy is: `V = 0.5 * x.T @ K @ x ` where `x = (x1, x2)`, the coordinates of the two atoms.

### Constructing a stiffness/force-constant matrix

```
import numpy as np


k = 2450.0
K = np.array([[ k, -k],
[-k, k]])


vals, vecs = np.linalg.eigh(K)
print('Eigenvalues:', vals)
print('Eigenvectors (columns):\n', vecs)


# Frequencies ignoring mass (omega = sqrt(lambda))
omega = np.sqrt(vals)
print('omega:', omega)


# Show orthogonality of eigenvector matrix U
U = vecs
print('U.T @ U close to identity?:', np.allclose(U.T @ U, np.eye(2)))
print('U @ U.T close to identity?:', np.allclose(U @ U.T, np.eye(2)))


# Check diagonalization: Lambda = U.T @ K @ U
Lambda = U.T @ K @ U
print('Lambda (should be diagonal with eigenvalues):\n', Lambda)
print('Diagonal equals vals?:', np.allclose(np.diag(Lambda), vals))
print('Max off-diagonal (round-off):', np.max(np.abs(Lambda - np.diag(np.diag(Lambda)))))


# Interpretation of modes:
# For vals approx [0, 2k], the eigenvector for 0 is [1,1] (translation mode),
# and the eigenvector for 2k is [1,-1] (stretch/compression, relative motion).
```

```
k = 2450 # the force constant in an appropriate [energy]/[length] units
K = np.array([[ k,-k]
             ,[-k, k]])
```

### Diagonalize to find frequencies and normal mode displacement patterns

```
vals, vecs = np.linalg.eigh(K)
vals, vecs
```

### Vibrational frequencies (ignoring mass)

```
omega = np.sqrt(vals)
omega
```
